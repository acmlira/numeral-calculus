```python
def F(x):
    return ((x ** 2) - 1)
```

```python
def signal_check(Fa, Fb):
    if(Fa * Fb < 0):
        return False
    return True
```

```python
def E_check(a, b, E):
    if((b - a) < E):
        return True
    return False
```

```python
def mean(a, b):
    return (a + b)/2
```

```python
def bissection(a, b, E, maxIter = 50):
    
    Fa = F(a)
    Fb = F(b)
    
    if(signal_check(Fa, Fb)):
        return False
    
    if(E_check(a, b, E)):
        return False
    
    x = mean(a, b)
    Fx = F(x)
    
    print("k\t  a\t\t  Fa\t\t  b\t\t  Fb\t\t  x\t\t  Fx\t\tAproximação")
    k = 1
    while k < maxIter:
        k += 1
        
        if(signal_check(Fa, Fx)):
            a = x
            Fa = Fx
        else:
            b = x
            Fb = Fx
        x = mean(a, b)
        Fx = F(x)
        print("-\t%e\t%e\t%e\t%e\t%e\t%e\t%e" % (a, Fa, b, Fb, x, Fx, E_check(a, b, E)))
    
    return x
```

```python
bissection(0.5, 1.25, 0.001)
```
